```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
lambdas = 0.3
gamma = 0.1

def SIR(t,z):
    S,I,R = z
    dSdt = -lambdas*S*I
    dIdt = lambdas*S*I-gamma*I
    dRdt = gamma*I
    return [dSdt, dIdt, dRdt]

S0 = 1
I0 = 20/2500
R0 = 0
Z0 = [S0,I0,R0]

t_span = (0,200)
t_eval = np.linspace(0, 200, 200)

sol = solve_ivp(SIR, t_span=t_span, y0=Z0, t_eval=t_eval, method='RK45')

sigma = lambdas/gamma
print(sigma)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(sol.t,sol.y[0], label="S")
plt.plot(sol.t,sol.y[1], label="I")
plt.plot(sol.t,sol.y[2], label="R")
plt.xlabel("Czas")
plt.ylabel("ludzie")
plt.legend()
plt.title("MODEL SIR")
plt.grid()
plt.show()
```
